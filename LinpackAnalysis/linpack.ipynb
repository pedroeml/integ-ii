```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tsv_files = ['LINPACK/Native/6000.tsv', 'LINPACK/LXC/6000.tsv', 'LINPACK/KVM/6000.tsv', 'LINPACK/Docker/6000.tsv']
```

```python
def get_dataframe(tsv_file):
    return pd.read_csv(tsv_file, sep='\s+')
```

# The collected data

## Native

```python
get_dataframe(tsv_files[0])
```

## LXC

```python
get_dataframe(tsv_files[1])
```

## KVM

```python
get_dataframe(tsv_files[2])
```

## Docker

```python
get_dataframe(tsv_files[3])
```

# Joining the data

```python
dfs = [pd.read_csv(tsv_files[i], sep='\s+') for i in range(len(tsv_files))]
dfs_gflops = [df['GFlops'] for df in dfs]
platforms = ['Native', 'LXC', 'KVM', 'Docker']
data = {}

for i in range(len(tsv_files)):
    platform = platforms[i]
    gflops = dfs_gflops[i]
    data[platform] = gflops

df = pd.DataFrame(data=data)
df
```

# Data visualization

```python
plt.figure()
ax = df.plot(title='Linpack Benchmark')
plt.xlabel('Test nº')
plt.ylabel('GFlops')
plt.savefig('LINPACK_6000.png')
plt.show()
```

# Data analysis

```python
d = {}
for key in platforms:
    min = df[key].min()
    avg = df[key].mean()
    max = df[key].max()
    std = df[key].std()
    d[key] = [min, avg, max, std]
d
```

```python
df_analysis = pd.DataFrame(data=d, index=['Minimum', 'Average', 'Maximum', 'Standard deviation'])
df_analysis
```

## Data comparison

```python
plt.figure()
ax = df_analysis.plot.bar(title='Linpack Benchmark comparison')
plt.ylabel('GFlops')
plt.tight_layout()
plt.savefig('LINPACK_6000_barplot.png')
plt.show()
```

```python
ax = df.boxplot()
ax.set_ylim([51, 59])
plt.title('Linpack Benchmark comparison')
plt.ylabel('GFlops')
plt.savefig('LINPACK_6000_boxplot.png')
plt.show()
```
